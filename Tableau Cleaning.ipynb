```python
import pandas as pd
import numpy as np
```

```python
file1 = pd.read_csv("JC-201906-citibike-tripdata.csv")
file2 = pd.read_csv("JC-202006-citibike-tripdata.csv")
```

```python
file1.head()
```

```python
file2.head()
```

```python
file1['gender'] = file1['gender'].astype(str)
file2['gender'] = file2['gender'].astype(str)
gender_2019 = file1['gender']
changed_2019_male = gender_2019.replace('1', 'Male')
changed_2019_female = changed_2019_male.replace('2', 'Female')
changed_2019_unknown = changed_2019_female.replace('0', 'Unknown')
file1['gender'] = changed_2019_unknown
file1
```

```python
gender_2020 = file2['gender']
changed_2020_male = gender_2020.replace('1', 'Male')
changed_2020_female = changed_2020_male.replace('2', 'Female')
changed_2020_unknown = changed_2020_female.replace('0', 'Unknown')
file2['gender'] = changed_2020_unknown
file2
```

```python
current_year = 2023

birth_year_2019 = file1['birth year']
age = current_year - birth_year_2019
file1['age'] = age

birth_year_2020 = file2['birth year']
age = current_year - birth_year_2020
file2['age'] = age
```

```python
file1.to_csv("06_2019_citibike_tripdata_cleaned.csv", index=False, header=True)
file2.to_csv("06_2020_citibike_tripdata_cleaned.csv", index=False, header=True)
```
